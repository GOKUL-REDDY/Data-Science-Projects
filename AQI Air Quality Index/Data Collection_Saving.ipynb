```python
from bs4 import BeautifulSoup
import os
import csv
import sys
```

```python
def met_data(month, year):
    
    file_html = open('Data/Html_Data/{}/{}.html'.format(year,month), 'rb')
    plain_text = file_html.read()
    #plain text will comtain all the data of the respective html file>>ALL TAGS

    tempD = []
    finalD = []

    soup = BeautifulSoup(plain_text, "lxml")
    #lxml="Format of beautifulSoup"
    for table in soup.findAll('table', {'class': 'medias mensuales numspan'}):
        #table an class is you can find it in the html code
        for tbody in table:
            #Basically tbody is tag
            for tr in tbody:
                #here its tr is the tag
                a = tr.get_text()
                #its gets the whole ROW.now we append that 15 to list tempD
                tempD.append(a)

    rows = len(tempD) / 15
    #we are dividing with 15 because its hasving that many only!

    for times in range(round(rows)):
        newtempD = []
        #this is used for pariticula data that is individual 15 values
        for i in range(15):
            newtempD.append(tempD[0])
            tempD.pop(0)
            #pop says is removing the previous one,....so its will be FinalD.
        finalD.append(newtempD)

    length = len(finalD)

    finalD.pop(length - 1)
    finalD.pop(0)
#Dropping all the useless features
    for a in range(len(finalD)):
        finalD[a].pop(6)
        finalD[a].pop(13)
        finalD[a].pop(12)
        finalD[a].pop(11)
        finalD[a].pop(10)
        finalD[a].pop(9)
        finalD[a].pop(0)

    return finalD
#SO FINALLY WE WILL BE HAVING 13-18 YEARS OF DATA IN finalD RESPECTIVELY
def data_combine(year, cs):
    for a in pd.read_csv('Data/Real-Data/real_' + str(year) + '.csv', chunksize=cs):
        df = pd.DataFrame(data=a)
        mylist = df.values.tolist()
    return mylist


```

```python
if __name__ == "__main__":
    if not os.path.exists("Data/Real-Data"):
        os.makedirs("Data/Real-Data")
    for year in range(2013, 2017):
        final_data = []
        with open('Data/Real-Data/real_' + str(year) + '.csv', 'w') as csvfile:
            wr = csv.writer(csvfile, dialect='excel')
            wr.writerow(
                ['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM', 'PM 2.5'])
        for month in range(1, 13):
            temp = met_data(month, year)
            final_data = final_data + temp
        #---------------Till here we got the values of AQI Now Its time to get the independent feartue
        
            
        pm = getattr(sys.modules[__name__], 'avg_data_{}'.format(year))()

        if len(pm) == 364:
            pm.insert(364, '-')

        for i in range(len(final_data)-1):
            # final[i].insert(0, i + 1)
            final_data[i].insert(8, pm[i])

        with open('Data/Real-Data/real_' + str(year) + '.csv', 'a') as csvfile:
            wr = csv.writer(csvfile, dialect='excel')
            for row in final_data:
                flag = 0
                for elem in row:
                    if elem == "" or elem == "-":
                        flag = 1
                if flag != 1:
                    wr.writerow(row)
                    
    data_2013 = data_combine(2013, 600)
    data_2014 = data_combine(2014, 600)
    data_2015 = data_combine(2015, 600)
    data_2016 = data_combine(2016, 600)
     
    total=data_2013+data_2014+data_2015+data_2016
    
    with open('Data/Real-Data/Real_Combine.csv', 'w') as csvfile:
        wr = csv.writer(csvfile, dialect='excel')
        wr.writerow(
            ['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM', 'PM 2.5'])
        wr.writerows(total)
        
        
df=pd.read_csv('Data/Real-Data/Real_Combine.csv')
```
